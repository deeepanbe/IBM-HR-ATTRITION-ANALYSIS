```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\dell\Desktop\PROJECT\IBM HR ANALYST\data\WA_Fn-UseC_-HR-Employee-Attrition.csv')


# Show the first 5 rows of the data
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Create a countplot for Attrition
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition')

# Add title
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Count")

# Show the chart
plt.show()
```

```python
# Save the chart as a PNG image
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition')
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Count")

# Save to outputs folder
plt.savefig(r'C:\Users\dell\Desktop\PROJECT\IBM HR ANALYST\outputs/attrition_chart.png')
plt.close()
```

```python
# Save Age vs Attrition chart
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Attrition', kde=True, bins=20)

plt.title("Attrition by Age")
plt.xlabel("Employee Age")
plt.ylabel("Number of Employees")
plt.legend(title='Attrition')
plt.tight_layout()

# Save to outputs folder
plt.savefig(r'C:\Users\dell\Desktop\PROJECT\IBM HR ANALYST\outputs/age_vs_attrition.png')


plt.close()
```

```python
# Save Attrition vs Salary chart
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')

plt.title("Attrition vs Monthly Income")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.tight_layout()

# Save the chart to your outputs folder
plt.savefig(r'C:\Users\dell\Desktop\PROJECT\IBM HR ANALYST\outputs\salary_vs_attrition.png')
plt.close()
```
